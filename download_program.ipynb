```python
# # function that take all the pdf and convert them to a text
# from PyPDF2 import PdfReader
# import textract
# def pdf_to_text(program_number,generated_url):
   
#     # from nltk.tokenize import word_tokenize
#     # from nltk.corpus import stopwords
#     # import nltk
#     # nltk.download('punkt')
#     # nltk.download('stopwords')
#     for generated_url in generated_urls:
#           # count += 1
#           url = unquote(generated_url)
#           if "הוראות התכנית" in url:
#             url = unquote(generated_url)
#             destination = f"out/{program_number}/{url.split('fn=')[-1].split('&')[0]}"
#             pdffileobj=open(destination,'rb')
#             pdfReader=PdfReader(pdffileobj)
#             num_pages = len(pdfReader.pages)
#             count = 0
#             text = ""
#             while count < num_pages:
#                 pageObj = pdfReader.pages[count]
#                 count +=1
#                 text += pageObj.extract_text()
#             if text != "":
#                 text = text
#             else:
#                 text = textract.process(destination, method='tesseract', language='eng')
#     return text

# import re

# text = pdf_to_text(program_number,generated_url)

# def remove_unwanted_text(text):
#     # Define the pattern for the unwanted text
#     unwanted_pattern = r'תכנון זמין\s+[0-9]\s+מונה\s+הדפסה'

#     # Use re.sub to replace occurrences of the unwanted pattern with an empty string
#     cleaned_text = re.sub(unwanted_pattern, '', text)


#     # Remove empty lines
#     cleaned_text = re.sub(r'\n\s*\n', '\n', cleaned_text)

#     return cleaned_text.strip()  # Optional: Strip leading and trailing spaces

# # Assuming 'text' is the extracted text from your PDF
# cleaned_text = remove_unwanted_text(text)
# print(cleaned_text)


from urllib.parse import unquote
from PyPDF2 import PdfReader
import textract
import re

def pdf_to_text(program_number, generated_urls):
    texts = []
    for generated_url in generated_urls:
        url = unquote(generated_url)
        if "הוראות התכנית" in url:
            url = unquote(generated_url)
            destination = f"out/{program_number}/{url.split('fn=')[-1].split('&')[0]}"
            pdffileobj = open(destination, 'rb')
            pdfReader = PdfReader(pdffileobj)
            num_pages = len(pdfReader.pages)
            count = 0
            text = ""
            while count < num_pages:
                pageObj = pdfReader.pages[count]
                count += 1
                text += pageObj.extract_text()
            if text != "":
                texts.append(text)
            else:
                text = textract.process(destination, method='tesseract', language='eng')
                texts.append(text)
    return texts

text = pdf_to_text(program_number, generated_urls)

def remove_unwanted_text(text):
    # Define the pattern for the unwanted text
    unwanted_pattern = r'תכנון זמין\s+[0-9]\s+מונה\s+הדפסה'

    # Use re.split to split the text based on the unwanted pattern
    chunks = re.split(unwanted_pattern, text[0])

    # Remove empty lines in each chunk and filter out empty chunks
    chunks = [re.sub(r'\n\s*\n', '\n', chunk).strip() for chunk in chunks if chunk.strip()]

    return chunks

cleaned_texts = remove_unwanted_text(text)

# Print or process each non-empty chunk separately
for idx, cleaned_text in enumerate(cleaned_texts):
    print(f"Chunk {idx + 1}:\n{cleaned_text}\n")    
# print("start",cleaned_texts[0])    

```
