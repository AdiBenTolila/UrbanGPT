````python
# # function that take all the pdf and convert them to a text
# from PyPDF2 import PdfReader
# import textract
# def pdf_to_text(program_number,generated_url):
   
#     # from nltk.tokenize import word_tokenize
#     # from nltk.corpus import stopwords
#     # import nltk
#     # nltk.download('punkt')
#     # nltk.download('stopwords')
#     for generated_url in generated_urls:
#           # count += 1
#           url = unquote(generated_url)
#           if "הוראות התכנית" in url:
#             url = unquote(generated_url)
#             destination = f"out/{program_number}/{url.split('fn=')[-1].split('&')[0]}"
#             pdffileobj=open(destination,'rb')
#             pdfReader=PdfReader(pdffileobj)
#             num_pages = len(pdfReader.pages)
#             count = 0
#             text = ""
#             while count < num_pages:
#                 pageObj = pdfReader.pages[count]
#                 count +=1
#                 text += pageObj.extract_text()
#             if text != "":
#                 text = text
#             else:
#                 text = textract.process(destination, method='tesseract', language='eng')
#     return text

# import re

# text = pdf_to_text(program_number,generated_url)

# def remove_unwanted_text(text):
#     # Define the pattern for the unwanted text
#     unwanted_pattern = r'תכנון זמין\s+[0-9]\s+מונה\s+הדפסה'

#     # Use re.sub to replace occurrences of the unwanted pattern with an empty string
#     cleaned_text = re.sub(unwanted_pattern, '', text)


#     # Remove empty lines
#     cleaned_text = re.sub(r'\n\s*\n', '\n', cleaned_text)

#     return cleaned_text.strip()  # Optional: Strip leading and trailing spaces

# # Assuming 'text' is the extracted text from your PDF
# cleaned_text = remove_unwanted_text(text)
# print(cleaned_text)


from urllib.parse import unquote
from PyPDF2 import PdfReader
import textract
import re

def pdf_to_text(program_number, generated_urls):
    texts = []
    for generated_url in generated_urls:
        url = unquote(generated_url)
        if "הוראות התכנית" in url:
            url = unquote(generated_url)
            destination = f"out/{program_number}/{url.split('fn=')[-1].split('&')[0]}"
            pdffileobj = open(destination, 'rb')
            pdfReader = PdfReader(pdffileobj)
            num_pages = len(pdfReader.pages)
            count = 0
            text = ""
            while count < num_pages:
                pageObj = pdfReader.pages[count]
                count += 1
                text += pageObj.extract_text()
            if text != "":
                texts.append(text)
            else:
                text = textract.process(destination, method='tesseract', language='eng')
                texts.append(text)
    return texts

text = pdf_to_text(program_number, generated_urls)

def remove_unwanted_text(text):
    # Define the pattern for the unwanted text
    unwanted_pattern = r'תכנון זמין\s+[0-9]\s+מונה\s+הדפסה'

    # Use re.split to split the text based on the unwanted pattern
    chunks = re.split(unwanted_pattern, text[0])

    # Remove empty lines in each chunk and filter out empty chunks
    chunks = [re.sub(r'\n\s*\n', '\n', chunk).strip() for chunk in chunks if chunk.strip()]

    return chunks

cleaned_texts = remove_unwanted_text(text)

# Print or process each non-empty chunk separately
for idx, cleaned_text in enumerate(cleaned_texts):
    print(f"Chunk {idx + 1}:\n{cleaned_text}\n")    
# print("start",cleaned_texts[0])    

````

````python
# from translate import Translator

# def split_text(text, max_length=500):
#     chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]
#     return chunks

# translator = Translator(to_lang="EN", from_lang="HE")
# translated_chunks = []

# for i, original_chunk in enumerate(cleaned_texts):
#     text_chunks = split_text(original_chunk)
#     chunk_translations = []

#     for chunk in text_chunks:
#         translation = translator.translate(chunk)
#         chunk_translations.append(translation)

#     # Concatenate translations for each original chunk
#     translated_chunk = ' '.join(chunk_translations)
#     translated_chunks.append(translated_chunk)

# # Print the results
# for idx, translated_chunk in enumerate(translated_chunks):
#     print(f"Original Chunk {idx + 1}:\n{cleaned_texts[idx]}\n")
#     print(f"Translated Chunk {idx + 1}:\n{translated_chunk}\n")

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

# Define the model and tokenizer names
model_name = "Helsinki-NLP/opus-mt-tc-big-he-en"

# Create tokenizer and model instances
tokenizer = AutoTokenizer.from_pretrained(model_name, cache_dir="cacheModels", force_download=True)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name, cache_dir="cacheModels", force_download=True)

# Create translation pipeline using the tokenizer and model
translate_pipe = pipeline("translation", model=model, tokenizer=tokenizer)

# Split the text into chunks at every dot
def split_text_at_dots(text):
    return text.split('.')

# Translate each chunk
translated_chunks = []

for i, original_chunk in enumerate(cleaned_texts):
    text_chunks = split_text_at_dots(original_chunk)
    chunk_translations = []

    for chunk in text_chunks:
        translate = translate_pipe(chunk)
        chunk_translations.append(translate[0]['translation_text'])

    # Join translations for each original chunk using a dot
    translated_chunk = '.'.join(chunk_translations)
    translated_chunks.append(translated_chunk)

# Print the results
for idx, translated_chunk in enumerate(translated_chunks):
    print(f"Original Chunk {idx + 1}:\n{cleaned_texts[idx]}\n")
    print(f"Translated Chunk {idx + 1}:\n{translated_chunk}\n")
````

````python
from langchain.document_loaders import HuggingFaceDatasetLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
from transformers import AutoTokenizer, pipeline
from langchain import HuggingFacePipeline
from langchain.chains import RetrievalQA
````

````python
# Define the path to the pre-trained model you want to use
modelPath = "sentence-transformers/all-MiniLM-l6-v2"

# Create a dictionary with model configuration options, specifying to use the CPU for computations
model_kwargs = {'device':'cpu'}

# Create a dictionary with encoding options, specifically setting 'normalize_embeddings' to False
encode_kwargs = {'normalize_embeddings': False}

# Initialize an instance of HuggingFaceEmbeddings with the specified parameters
embeddings = HuggingFaceEmbeddings(
    model_name=modelPath,     # Provide the pre-trained model's path
    model_kwargs=model_kwargs, # Pass the model configuration options
    encode_kwargs=encode_kwargs # Pass the encoding options
)
````

````python
from langchain.docstore.document import Document

documents = [Document(page_content=chunk,metadata={"source": "local"}) for chunk in translated_chunks]
db = FAISS.from_documents(documents, embeddings)

question = "In the construction planning document provided, identify and extract the street or specific location where the construction is planned. If the street or location is Uruguay answer yes, if not answer no and if the document does not contain this information, please indicate that it does not exist"
searchDocs = db.similarity_search(question)
print(searchDocs[0])
````

````python
# Callbacks support token-wise streaming
callback_manager = CallbackManager([StreamingStdOutCallbackHandler()])


llm = LlamaCpp(
    # model_path="models/leo-hessianai-7b.Q4_K_M.gguf",
    model_path="../../mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    temperature=0.75,
    max_tokens=2000,
    top_p=1,
    callback_manager=callback_manager, 
    n_ctx=4096,  # Reduce the context size
    verbose=True,
)
````
